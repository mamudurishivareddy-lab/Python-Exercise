Question 7: Given below is a tuple. Sort the tuple and print the result.

Question 9: Given below are two sets, perform the operations given below:
1. Print their union
2. Print their intersection
3. Print their difference
4. Print their symmetric difference

```python
# Given below are two sets:
a={1,2,3,5,6,7}
b={4,5,8,9,2,3}

#Yes — but not directly, and this is where people get sloppy and write bad code.
#Set operations do NOT work on lists. They work on sets. You must convert first.

# Print their union:All unique elements in both sets.         
c=set(a)|set(b) #print(a.union(b))
print(c)
# Print their intersection:Common elements in both sets. 
d=set(a)&set(b) #print(a.intersection(b))
print(d)
# Print their difference:Elements in A that are not in B                 
e=set(a)-set(b) #print(a.difference(b))
print(e)
# Print their symmetric difference: Elements in A or B but not in both
f=set(a)^set(b)
print(f)
```
